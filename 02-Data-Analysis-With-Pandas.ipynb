## Change Display Settings

By default, Pandas will display 60 rows and 20 columns. I often change [Pandas' default display settings](https://pandas.pydata.org/pandas-docs/stable/user_guide/options.html) to show more rows or columns.

```python
pd.options.display.max_rows = 200
```

## Get Data

```python
seattle_df = pd.read_csv('Seattle-Library_2020-2021.csv')
```

```python
seattle_df.head()
```

## Groupby

```python
seattle_df.groupby('Title')[['Checkouts']].sum()
```

```python
seattle_df.groupby('Title')[['Checkouts']].sum().sort_values(by='Checkouts',ascending=False)[:50]
```

```python
seattle_df['PublicationYear'].value_counts()
```

```python
seattle_df.groupby(['Title', 'PublicationYear'])[['Checkouts']].sum().sort_values(by=['PublicationYear'],ascending=[True])[:50]
```

```python
seattle_df.columns
```

```python
seattle_df.groupby('CheckoutMonthYear')[['Checkouts']].sum()
```

```python
seattle_df.groupby(['Title', 'CheckoutMonthYear'])[['Checkouts']].sum()
```

## Reset Index

```python
seattle_df.groupby(['Title', 'CheckoutMonthYear'])[['Checkouts']].sum().reset_index()
```

## Plot

```python
seattle_df.groupby('CheckoutMonthYear')['Checkouts'].sum().plot()
```

```python
seattle_df['MaterialType'].value_counts().plot(kind='bar')
```

```python
seattle_df[seattle_df['Title'] == 'So You Want to Talk about Race (Unabridged)'].plot(x='CheckoutMonthYear', y='Checkouts')
```

```python
seattle_df.plot(x='CheckoutMonthYear', y='Checkouts')
```

```python
title_filter = seattle_df['Title'].str.contains('Frog and Toad')
seattle_df[title_filter].plot(x='CheckoutMonthYear', y='Checkouts')
```
